# Code Python

Dans ce notebook, nous allons simuler l'évolution de la vitesse d'un véhicule pendant le freinage en tenant compte de la masse du véhicule, du coefficient de frottement et de la gravité. Nous allons utiliser la formule suivante pour calculer la décélération :

a = (µ * m * g) / m

où µ est le coefficient de frottement, m est la masse du véhicule et g est l'accélération due à la gravité.


Nous allons maintenant définir les paramètres pour notre simulation et utiliser la fonction `calculate_velocity_with_mass` pour calculer l'évolution de la vitesse du véhicule pendant le freinage. Les paramètres sont les suivants :

- m : masse du véhicule (kg)
- v0 : vitesse initiale (m/s)
- µ : coefficient de frottement
- g : accélération due à la gravité (m/s²)
- dt : intervalle de temps entre les mesures (s)


```python
# Paramètres
m = 1200  # Masse du véhicule en kg
v0 = 27.7  # Vitesse initiale en m/s (environ 180 km/h)
µ = 0.85  # Coefficient de frottement
g = 9.81  # Gravité en m/s²
dt = 0.01  # Intervalle de temps
```

Ici, on a la fonction qui calcule la vitesse du véhicule à chaque intervalle de temps:

```python
import matplotlib.pyplot as plt

def calculate_velocity(v0, m, µ, g, dt):
    Fn = m * g  # Force normale
    Ff = µ * Fn  # Force de frottement
    a = Ff / m  # Décélération
    v = [v0]
    t = [0]
    while v[-1] > 0:
        v_new = max(v[-1] - a * dt, 0)  # S'assurer que la vitesse ne devient pas négative
        v.append(v_new)
        t.append(t[-1] + dt)
    return t, v, a
```

Ici, on a une fonction qui permet de nous donner la vitesse exacte du véhciule à un instant donné:

```python
def exact_velocity(v0, a, t):
    v = max(v0 - a * t, 0)
    return v
```

Voici l'affichage du graphique selon les calculs:

```python
# Calcul de l'évolution de la vitesse avec prise en compte de la masse du véhicule
t, v, a = calculate_velocity(v0, m, µ, g, dt)

# Création du graphique
plt.plot(t, v)
plt.xlabel('Temps (s)')
plt.ylabel('Vitesse (m/s)')
plt.title("Évolution de la vitesse pendant le freinage en tenant compte de la masse du véhicule")
plt.grid(True)
plt.show()
```

```python
# Affichage du résultat exact
time_to_stop = t[-1]
final_velocity = v[-1]
print(f"Le véhicule s'arrête en {time_to_stop:.2f} secondes avec une vitesse finale de {final_velocity:.2f} m/s.")
```

```python
# Calcul de la vitesse exacte à un moment t spécifique
t_specific = 2.5  # Temps spécifique en secondes
v_specific = exact_velocity(v0, a, t_specific)
print(f"À t = {t_specific} secondes, la vitesse du véhicule est de {v_specific:.2f} m/s.")
```
